```
with open('stuff1.txt', 'r', encoding="utf-8") as file:
  data = file.read().lower()
print(type(data))
print(data)

data = data.splitlines()
```

```
print(data)
```

```
#new_data = [ ]
for values in data:
  try:
    new_data.append(values.split('- ')[1])
  except:
    pass
```

```
print(new_data)
```

```
for stuff in new_data:

  with open('input.txt', mode='a', encoding='utf-8') as f:
    f.write(stuff)
    f.write('\n')
f.close()
```
